```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
# function to remove stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)
```

```python
df['description'] = df['description'].apply(lambda x: remove_stopwords(x))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def freq_words(x, terms = 30): 
  all_words = ' '.join([text for text in x]) 
  all_words = all_words.split() 
  fdist = nltk.FreqDist(all_words) 
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
  
  # selecting top n(terms) most frequent words 
  d = words_df.nlargest(columns="count", n = terms) 
  #print(d['word'])
  
  for wor in d['word']:
        if( not (wor==("women's") or wor==("men's") ) ):
            stop_words.add(wor)
  #stop_words.add(d['word'])
  #print(stop_words)
  
  # visualize words and frequencies
  plt.figure(figsize=(12,15)) 
  ax = sns.barplot(data=d, x= "count", y = "word") 
  ax.set(ylabel = 'Word') 
  plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

```python
tfidf = TfidfVectorizer(max_features = 20000, ngram_range = (1,3), analyzer ='char')
```

Now we will train different models for different branches of the category tree, The intution behind this is to increase the acceracy of the prediction However one of the major drawback of this approach is the overhead created in training and prediction.

```python
dict = {}
for ind in df.index:
    cate = df['product_category_tree'][ind]
    #print(cate)
    for i in range(len(cate)):
        if cate[i ] not in  dict.keys():
            dict[cate[i]] = pd.DataFrame(columns = ['category', 'description'])
            dict[cate[i]] = dict.get(cate[i]).append({'category' : "Null" , 'description':df['description'][ind]}, ignore_index = True)
        if i == len(cate)-1 :
            dict[cate[i]] = dict.get(cate[i]).append({'category' : "Null2" , 'description':df['description'][ind]+"TEST"}, ignore_index = True)
        else:
            dict[cate[i]] = dict[cate[i]].append({'category': cate[i+1], 'description':df['description'][ind]}, ignore_index = True)
```

Now we will create indivual models for each category. The idea behind the prediction would be:$$ $$
    $step 1:$ Take the description and predict the root category.$$ $$
    $step 2:$ Once the root category is obtained use the root category model to predict the sub category.$$ $$
    $step 3:$ Now once the sub category is obtained, make it root and continue again from step 1.$$ $$
    $step 4:$ Repeat this until no subcategory is remaining.$$ $$
    

```python
dict['root'] = pd.DataFrame(columns = ['category', 'description'])
for ind in df.index:
    cate = df['product_category_tree'][ind]
    dict['root']= dict['root'].append({'category': cate[0], 'description':df['description'][ind]}, ignore_index = True)
    
```

# Creating category models and storing them in a dict called model_dict

```python
#Here we will avoid creating models where data available for a particular
# class is less that a particular number to avoid overheads
model_dict ={}
for cate in dict.keys():
    #print(1)
    #print(dict[cate]['description'])
    print(dict[cate].shape[0], dict[cate].shape[1])
    if dict[cate].shape[0] <= 6:
        continue
    if len(dict[cate]['category']) < 6 or len(dict[cate]['description'])<6:
        continue
    model_dict[cate] = LinearSVC()
    X = tfidf.fit_transform(dict[cate]['description'])
    Y = dict[cate]['category']
    model_dict[cate].fit(X, Y)
```
